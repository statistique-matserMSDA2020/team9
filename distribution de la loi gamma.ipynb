```python
### khoudia mbodji###
###distribution de la loi gamma###
###QUESTION 1 q3###
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma
gamma_distribution = gamma(loc = 10, scale = 10, a = 5)
x = np.linspace(-10, 40, 10000)
figure, ax = plt.subplots(1, 1)
ax.plot(x, gamma_distribution.pdf(x), 'g', lw=6)
print(x)
plt.title('estimation dun echantillon  de taille 10000 suivant une loi gamma Î³(10,.5).')
plt.show()
print(gamma_distribution)
```
